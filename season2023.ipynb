```python
# ============================================================
# Описательные статистики сезона 2023
# ============================================================

print(f"{'='*65}")
print(f"ОСНОВНЫЕ МЕТРИКИ НА УРОВНЕ ЗАПИСИ (тройка × матч):")
print(f"{'='*65}")

key_cols = ['icetime', 'CF', 'CA', 'FF', 'FA', 'GF', 'GA', 
            'xGF', 'xGA', 'shotAttemptsFor', 'shotAttemptsAgainst',
            'shotsOnGoalFor', 'shotsOnGoalAgainst',
            'highDangerShotsFor', 'highDangerShotsAgainst']
print(df[key_cols].describe().round(2).to_string())

# ============================================================
# Агрегация по тройкам
# ============================================================

agg = df.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    FF=('FF', 'sum'), FA=('FA', 'sum'),
    GF=('GF', 'sum'), GA=('GA', 'sum'),
    xGF=('xGF', 'sum'), xGA=('xGA', 'sum'),
    icetime=('icetime', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
    team=('playerTeam', 'first'),
    highDangerFor=('highDangerShotsFor', 'sum'),
    highDangerAgainst=('highDangerShotsAgainst', 'sum'),
).reset_index()

agg['cf_pct'] = agg['CF'] / (agg['CF'] + agg['CA'])
agg['ff_pct'] = agg['FF'] / (agg['FF'] + agg['FA'])
agg['xgf_pct'] = agg['xGF'] / (agg['xGF'] + agg['xGA'])
agg['gf_pct'] = agg['GF'] / (agg['GF'] + agg['GA'])
agg.loc[(agg['GF'] + agg['GA']) == 0, 'gf_pct'] = 0.5
agg['hd_pct'] = agg['highDangerFor'] / (agg['highDangerFor'] + agg['highDangerAgainst'])
agg.loc[(agg['highDangerFor'] + agg['highDangerAgainst']) == 0, 'hd_pct'] = 0.5

print(f"\n{'='*65}")
print(f"АГРЕГИРОВАННЫЕ ТРОЙКИ: {len(agg)}")
print(f"{'='*65}")
print(agg[['n_games', 'icetime', 'cf_pct', 'ff_pct', 
           'xgf_pct', 'gf_pct', 'hd_pct']].describe().round(3).to_string())

# ============================================================
# Распределение матчей на тройку
# ============================================================

print(f"\n{'='*65}")
print(f"РАСПРЕДЕЛЕНИЕ МАТЧЕЙ НА ТРОЙКУ:")
print(f"{'='*65}")
for lo, hi in [(5,7), (8,10), (11,15), (16,20), (21,30), (31,50), (51,82)]:
    n = ((agg['n_games'] >= lo) & (agg['n_games'] <= hi)).sum()
    pct = n / len(agg) * 100
    bar = '█' * int(pct / 2)
    print(f"  {lo:2d}-{hi:2d}: {n:4d} ({pct:5.1f}%) {bar}")

# ============================================================
# Статистика по игрокам
# ============================================================

print(f"\n{'='*65}")
print(f"СТАТИСТИКА ПО ИГРОКАМ:")
print(f"{'='*65}")

player_stats = {}
for _, row in df.iterrows():
    for p in [row['p1'], row['p2'], row['p3']]:
        if p not in player_stats:
            player_stats[p] = {'CF': 0, 'CA': 0, 'games': set(), 'lines': set()}
        player_stats[p]['CF'] += row['CF']
        player_stats[p]['CA'] += row['CA']
        player_stats[p]['games'].add(row['gameId'])
        player_stats[p]['lines'].add(row['lineId'])

pdf = pd.DataFrame([
    {'player': p, 
     'cf_pct': s['CF']/(s['CF']+s['CA']) if (s['CF']+s['CA'])>0 else 0.5,
     'n_games': len(s['games']), 
     'n_lines': len(s['lines'])}
    for p, s in player_stats.items()
])

print(f"  Всего игроков: {len(pdf)}")
print(f"  Матчей/игрок:  med={pdf['n_games'].median():.0f}, "
      f"mean={pdf['n_games'].mean():.1f}, "
      f"max={pdf['n_games'].max()}")
print(f"  Троек/игрок:   med={pdf['n_lines'].median():.0f}, "
      f"mean={pdf['n_lines'].mean():.1f}, "
      f"max={pdf['n_lines'].max()}")
print(f"  CF% игроков:   mean={pdf['cf_pct'].mean():.3f}, "
      f"std={pdf['cf_pct'].std():.3f}")

# ============================================================
# Корреляции метрик
# ============================================================

print(f"\n{'='*65}")
print(f"КОРРЕЛЯЦИИ МЕТРИК (агрегированные тройки):")
print(f"{'='*65}")

metrics = ['cf_pct', 'ff_pct', 'xgf_pct', 'gf_pct', 'hd_pct', 'n_games']
print(agg[metrics].corr().round(3).to_string())

# ============================================================
# Бинарная целевая переменная
# ============================================================

agg['y'] = (agg['cf_pct'] > 0.5).astype(int)

print(f"\n{'='*65}")
print(f"ЦЕЛЕВАЯ ПЕРЕМЕННАЯ (CF% > 50%):")
print(f"{'='*65}")
print(f"  y=1 (хорошие): {agg['y'].sum()} ({agg['y'].mean()*100:.1f}%)")
print(f"  y=0 (плохие):  {(1-agg['y']).sum():.0f} ({(1-agg['y'].mean())*100:.1f}%)")

# ============================================================
# Визуализация
# ============================================================

fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 1: Матчей на тройку
axes[0,0].hist(agg['n_games'], bins=40, color='steelblue', edgecolor='white')
axes[0,0].set_xlabel('Матчей')
axes[0,0].set_title(f'Матчей на тройку (n={len(agg)})\n'
                     f'med={agg["n_games"].median():.0f}, '
                     f'mean={agg["n_games"].mean():.1f}')

# 2: CF%
axes[0,1].hist(agg['cf_pct'], bins=35, color='steelblue', edgecolor='white')
axes[0,1].axvline(0.5, color='red', linestyle='--', linewidth=2)
axes[0,1].set_xlabel('CF%')
axes[0,1].set_title(f'CF% троек\n'
                     f'mean={agg["cf_pct"].mean():.3f}, '
                     f'std={agg["cf_pct"].std():.3f}')

# 3: CF% vs кол-во матчей
axes[0,2].scatter(agg['n_games'], agg['cf_pct'], alpha=0.4, s=15)
axes[0,2].axhline(0.5, color='red', linestyle='--')
axes[0,2].set_xlabel('Матчей')
axes[0,2].set_ylabel('CF%')
axes[0,2].set_title('CF% vs кол-во матчей\n(дисперсия уменьшается)')

# 4: CF% vs xGF%
r_val = agg['cf_pct'].corr(agg['xgf_pct'])
axes[1,0].scatter(agg['cf_pct'], agg['xgf_pct'], alpha=0.4, s=15)
axes[1,0].plot([0.3, 0.7], [0.3, 0.7], 'r--')
axes[1,0].set_xlabel('CF%')
axes[1,0].set_ylabel('xGF%')
axes[1,0].set_title(f'CF% vs xGF% (r={r_val:.3f})')

# 5: CF% игроков
axes[1,1].hist(pdf['cf_pct'], bins=35, color='coral', edgecolor='white')
axes[1,1].axvline(0.5, color='red', linestyle='--', linewidth=2)
axes[1,1].set_xlabel('CF% игрока')
axes[1,1].set_title(f'CF% игроков (n={len(pdf)})\n'
                     f'mean={pdf["cf_pct"].mean():.3f}')

# 6: Троек на игрока
axes[1,2].hist(pdf['n_lines'], bins=30, color='coral', edgecolor='white')
axes[1,2].set_xlabel('Кол-во троек')
axes[1,2].set_title(f'Троек на игрока\n'
                     f'med={pdf["n_lines"].median():.0f}')

plt.suptitle('Сезон 2023: описательные статистики', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('fig_s23_descriptive.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\n✅ Готово к моделированию!")
```

```python
import numpy as np
import pandas as pd
import teneva
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import reverse_cuthill_mckee
from sklearn.metrics import accuracy_score, roc_auc_score
import matplotlib.pyplot as plt
from time import perf_counter as tpc

# ============================================================
# Шаг 1: Загрузка и split по дате (75/25)
# ============================================================

df = pd.read_csv('season_2023_filtered.csv')
df['gameDate'] = pd.to_datetime(df['gameDate'])

all_dates = sorted(df['gameDate'].unique())
n_dates = len(all_dates)
split_date = all_dates[int(n_dates * 0.75)]

train_data = df[df['gameDate'] <= split_date].copy()
test_data = df[df['gameDate'] > split_date].copy()

print(f"{'='*65}")
print(f"SPLIT ПО ДАТЕ:")
print(f"{'='*65}")
print(f"  Всего игровых дней: {n_dates}")
print(f"  Split дата: {split_date.date()}")
print(f"  Train: {len(train_data):,} записей, "
      f"до {split_date.date()}")
print(f"  Test:  {len(test_data):,} записей, "
      f"после {split_date.date()}")

# ============================================================
# Шаг 2: Агрегация train по тройкам
# ============================================================

train_agg = train_data.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
).reset_index()

train_agg = train_agg[train_agg['n_games'] >= 3].reset_index(drop=True)
train_agg['cf_pct'] = train_agg['CF'] / (train_agg['CF'] + train_agg['CA'])
train_agg['y'] = (train_agg['cf_pct'] > 0.5).astype(float)

# Агрегация test по тройкам
test_agg = test_data.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
).reset_index()

test_agg = test_agg[test_agg['n_games'] >= 3].reset_index(drop=True)
test_agg['cf_pct'] = test_agg['CF'] / (test_agg['CF'] + test_agg['CA'])
test_agg['y'] = (test_agg['cf_pct'] > 0.5).astype(float)

print(f"\n  Train троек (≥3 матча): {len(train_agg)}, y=1: {train_agg['y'].mean():.3f}")
print(f"  Test троек (≥3 матча):  {len(test_agg)}, y=1: {test_agg['y'].mean():.3f}")

# ============================================================
# Шаг 3: Индексация игроков (из train)
# ============================================================

all_players = set()
for _, row in train_agg.iterrows():
    all_players.update([row['p1'], row['p2'], row['p3']])

# Считаем частоту каждого игрока в train
player_freq = {}
for _, row in train_data.iterrows():
    for p in [row['p1'], row['p2'], row['p3']]:
        player_freq[p] = player_freq.get(p, 0) + 1

# Сортируем по частоте (самый частый первый)
freq_sorted = sorted(all_players, key=lambda p: -player_freq.get(p, 0))
pidx = {pid: idx for idx, pid in enumerate(freq_sorted)}
n_pl = len(pidx)

print(f"\n  Игроков в train: {n_pl}")

# ============================================================
# Шаг 4: RCM-перестановка
# ============================================================

co = np.zeros((n_pl, n_pl), dtype=int)
for _, row in train_data.iterrows():
    players = [row['p1'], row['p2'], row['p3']]
    idxs = [pidx[p] for p in players if p in pidx]
    for i in range(len(idxs)):
        for j in range(i + 1, len(idxs)):
            co[idxs[i], idxs[j]] += 1
            co[idxs[j], idxs[i]] += 1

rcm = reverse_cuthill_mckee(csr_matrix(co))
print(f"  RCM перестановка: {len(rcm)} игроков")

# ============================================================
# Шаг 5: Бинарные матрицы
# ============================================================

# Train
I_trn = np.zeros((len(train_agg), n_pl), dtype=int)
y_trn = np.zeros(len(train_agg), dtype=float)

for i, (_, row) in enumerate(train_agg.iterrows()):
    for p in [row['p1'], row['p2'], row['p3']]:
        I_trn[i, pidx[p]] = 1
    y_trn[i] = row['y']

# Применяем RCM
I_trn = I_trn[:, rcm]

# Test: убираем тройки с неизвестными игроками
test_known = test_agg[
    test_agg['p1'].isin(pidx) &
    test_agg['p2'].isin(pidx) &
    test_agg['p3'].isin(pidx)
].reset_index(drop=True)

test_known['in_train'] = test_known['lineId'].isin(set(train_agg['lineId']))

I_tst = np.zeros((len(test_known), n_pl), dtype=int)
y_tst = np.zeros(len(test_known), dtype=float)

for i, (_, row) in enumerate(test_known.iterrows()):
    for p in [row['p1'], row['p2'], row['p3']]:
        I_tst[i, pidx[p]] = 1
    y_tst[i] = row['y']

# Применяем RCM
I_tst = I_tst[:, rcm]

# ============================================================
# Шаг 6: Убираем "мёртвые" столбцы
# ============================================================

good = [k for k in range(I_trn.shape[1]) if I_trn[:, k].max() == 1]
bad = [k for k in range(I_trn.shape[1]) if I_trn[:, k].max() == 0]

print(f"\n  Столбцов good: {len(good)}, bad: {len(bad)}")

if len(bad) > 0:
    # Убираем из test строки, где есть единица в bad столбце
    mask = np.ones(len(I_tst), dtype=bool)
    for k in bad:
        mask &= (I_tst[:, k] == 0)
    I_tst = I_tst[mask]
    y_tst = y_tst[mask]
    test_known = test_known[mask].reset_index(drop=True)
    print(f"  Убрано из test: {(~mask).sum()} строк")

I_trn = I_trn[:, good]
I_tst = I_tst[:, good]

print(f"\n{'='*65}")
print(f"ИТОГОВЫЕ ДАННЫЕ:")
print(f"{'='*65}")
print(f"  I_trn: {I_trn.shape}, y=1: {y_trn.mean():.3f}")
print(f"  I_tst: {I_tst.shape}, y=1: {y_tst.mean():.3f}")
print(f"  Новых троек в test: {(~test_known['in_train']).sum()} / {len(test_known)}")

# Проверка: каждая строка содержит ровно 3 единицы
assert (I_trn.sum(axis=1) == 3).all(), "Train: не все строки имеют 3 единицы!"
assert (I_tst.sum(axis=1) == 3).all(), "Test: не все строки имеют 3 единицы!"
print(f"  ✓ Каждая строка содержит ровно 3 единицы")

# ============================================================
# Шаг 7: ANOVA + ALS с разными рангами
# ============================================================

results = {}

for r in [2, 3, 5, 8]:
    print(f"\n{'='*65}")
    print(f"РАНГ r = {r}")
    print(f"{'='*65}")

    # --- ANOVA ---
    t = tpc()
    Y0 = teneva.anova(I_trn, y_trn, r=r, order=1)
    t_anova = tpc() - t

    y_p_trn_anova = teneva.get_many(Y0, I_trn)
    y_p_tst_anova = teneva.get_many(Y0, I_tst)

    trn_acc_anova = np.mean((y_p_trn_anova > 0.5) == y_trn)
    tst_acc_anova = np.mean((y_p_tst_anova > 0.5) == y_tst)

    print(f"  ANOVA: Train Acc={trn_acc_anova:.4f}, "
          f"Test Acc={tst_acc_anova:.4f} ({t_anova:.2f}s)")

    # --- ALS ---
    t = tpc()
    Y = teneva.als(
        I_trn, y_trn, Y0,
        nswp=30,
        I_vld=I_tst, y_vld=y_tst,
        log=True
    )
    t_als = tpc() - t

    y_p_trn = teneva.get_many(Y, I_trn)
    y_p_tst = teneva.get_many(Y, I_tst)

    trn_acc = np.mean((y_p_trn > 0.5) == y_trn)
    tst_acc = np.mean((y_p_tst > 0.5) == y_tst)
    trn_mae = np.mean(np.abs(y_p_trn - y_trn))
    tst_mae = np.mean(np.abs(y_p_tst - y_tst))

    try:
        tst_auc = roc_auc_score(y_tst, y_p_tst)
    except:
        tst_auc = 0.0

    # Знакомые vs новые
    known_mask = test_known['in_train'].values
    new_mask = ~known_mask

    acc_known = np.mean((y_p_tst[known_mask] > 0.5) == y_tst[known_mask]) if known_mask.sum() > 0 else 0
    acc_new = np.mean((y_p_tst[new_mask] > 0.5) == y_tst[new_mask]) if new_mask.sum() > 0 else 0

    print(f"\n  ALS:   Train Acc={trn_acc:.4f}, Test Acc={tst_acc:.4f} ({t_als:.2f}s)")
    print(f"  ALS:   Train MAE={trn_mae:.4f}, Test MAE={tst_mae:.4f}")
    print(f"  ALS:   Test AUC={tst_auc:.4f}")
    print(f"  ALS:   Знакомые Acc={acc_known:.4f} (n={known_mask.sum()}), "
          f"Новые Acc={acc_new:.4f} (n={new_mask.sum()})")

    results[r] = {
        'trn_acc_anova': trn_acc_anova, 'tst_acc_anova': tst_acc_anova,
        'trn_acc': trn_acc, 'tst_acc': tst_acc,
        'trn_mae': trn_mae, 'tst_mae': tst_mae,
        'tst_auc': tst_auc,
        'acc_known': acc_known, 'acc_new': acc_new,
        'y_pred': y_p_tst,
    }

# ============================================================
# Шаг 8: Итоговая таблица
# ============================================================

majority_acc = max(y_tst.mean(), 1 - y_tst.mean())

print(f"\n{'='*65}")
print(f"ИТОГОВАЯ ТАБЛИЦА:")
print(f"{'='*65}")
print(f"{'Ранг':>5} {'ANOVA':>8} {'ALS':>8} {'ALS':>8} {'ALS':>8} {'ALS':>8} "
      f"{'Знаком':>8} {'Новые':>8}")
print(f"{'':>5} {'Tst Acc':>8} {'Trn Acc':>8} {'Tst Acc':>8} {'Tst AUC':>8} "
      f"{'Tst MAE':>8} {'Acc':>8} {'Acc':>8}")
print(f"{'-'*65}")
print(f"{'Maj':>5} {'':>8} {'':>8} {majority_acc:>8.4f}")

for r, res in results.items():
    print(f"{r:>5} {res['tst_acc_anova']:>8.4f} {res['trn_acc']:>8.4f} "
          f"{res['tst_acc']:>8.4f} {res['tst_auc']:>8.4f} "
          f"{res['tst_mae']:>8.4f} {res['acc_known']:>8.4f} "
          f"{res['acc_new']:>8.4f}")

# ============================================================
# Шаг 9: Визуализация
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

ranks = list(results.keys())

# 1: Accuracy vs ранг
axes[0,0].plot(ranks, [results[r]['trn_acc'] for r in ranks],
               'o-', label='ALS Train', linewidth=2)
axes[0,0].plot(ranks, [results[r]['tst_acc'] for r in ranks],
               's-', label='ALS Test', linewidth=2)
axes[0,0].plot(ranks, [results[r]['tst_acc_anova'] for r in ranks],
               '^--', label='ANOVA Test', linewidth=2)
axes[0,0].axhline(majority_acc, color='red', linestyle='--',
                   label=f'Majority={majority_acc:.3f}')
axes[0,0].set_xlabel('Ранг')
axes[0,0].set_ylabel('Accuracy')
axes[0,0].set_title('Accuracy vs ранг')
axes[0,0].legend()

# 2: MAE vs ранг
axes[0,1].plot(ranks, [results[r]['trn_mae'] for r in ranks],
               'o-', label='Train MAE', linewidth=2)
axes[0,1].plot(ranks, [results[r]['tst_mae'] for r in ranks],
               's-', label='Test MAE', linewidth=2)
axes[0,1].set_xlabel('Ранг')
axes[0,1].set_ylabel('MAE')
axes[0,1].set_title('MAE vs ранг')
axes[0,1].legend()

# 3: Predicted vs Actual (лучший ранг)
best_r = max(results, key=lambda r: results[r]['tst_acc'])
y_best = results[best_r]['y_pred']
cf_actual = test_known['cf_pct'].values

axes[1,0].scatter(cf_actual, y_best, alpha=0.5, s=20,
                   c=test_known['in_train'].map({True: 'steelblue', False: 'coral'}))
axes[1,0].plot([0.3, 0.7], [0.3, 0.7], 'r--')
axes[1,0].axhline(0.5, color='gray', linestyle=':')
axes[1,0].axvline(0.5, color='gray', linestyle=':')
axes[1,0].set_xlabel('Actual CF%')
axes[1,0].set_ylabel('Predicted')
axes[1,0].set_title(f'Pred vs Actual (r={best_r})\n'
                     f'Синий=знакомые, Красный=новые')

# 4: Знакомые vs новые по рангам
axes[1,1].plot(ranks, [results[r]['acc_known'] for r in ranks],
               'o-', label='Знакомые', linewidth=2)
axes[1,1].plot(ranks, [results[r]['acc_new'] for r in ranks],
               's-', label='Новые', linewidth=2)
axes[1,1].axhline(majority_acc, color='red', linestyle='--',
                   label=f'Majority={majority_acc:.3f}')
axes[1,1].set_xlabel('Ранг')
axes[1,1].set_ylabel('Test Accuracy')
axes[1,1].set_title('Знакомые vs новые тройки')
axes[1,1].legend()

plt.suptitle('TT-ANOVA + ALS: сезон 2023 (бинарная матрица)',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('fig_tt_binary_results.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import pandas as pd
import teneva
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import reverse_cuthill_mckee
from sklearn.metrics import accuracy_score, roc_auc_score
import matplotlib.pyplot as plt
from time import perf_counter as tpc

# ============================================================
# Шаг 1: Загрузка и split по дате (75/25)
# ============================================================

df = pd.read_csv('season_2023_filtered.csv')
df['gameDate'] = pd.to_datetime(df['gameDate'])

all_dates = sorted(df['gameDate'].unique())
n_dates = len(all_dates)
split_date = all_dates[int(n_dates * 0.75)]

train_data = df[df['gameDate'] <= split_date].copy()
test_data = df[df['gameDate'] > split_date].copy()

print(f"{'='*65}")
print(f"SPLIT ПО ДАТЕ:")
print(f"{'='*65}")
print(f"  Train: {len(train_data):,} записей, до {split_date.date()}")
print(f"  Test:  {len(test_data):,} записей")

# ============================================================
# Шаг 2: Агрегация по тройкам
# ============================================================

train_agg = train_data.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
).reset_index()
train_agg = train_agg[train_agg['n_games'] >= 3].reset_index(drop=True)
train_agg['cf_pct'] = train_agg['CF'] / (train_agg['CF'] + train_agg['CA'])
train_agg['y'] = (train_agg['cf_pct'] > 0.5).astype(float)

test_agg = test_data.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
).reset_index()
test_agg = test_agg[test_agg['n_games'] >= 3].reset_index(drop=True)
test_agg['cf_pct'] = test_agg['CF'] / (test_agg['CF'] + test_agg['CA'])
test_agg['y'] = (test_agg['cf_pct'] > 0.5).astype(float)

print(f"  Train троек: {len(train_agg)}, y=1: {train_agg['y'].mean():.3f}")
print(f"  Test троек:  {len(test_agg)}, y=1: {test_agg['y'].mean():.3f}")

# ============================================================
# Шаг 3: Индексация + RCM
# ============================================================

all_players = set()
for _, row in train_agg.iterrows():
    all_players.update([row['p1'], row['p2'], row['p3']])

player_freq = {}
for _, row in train_data.iterrows():
    for p in [row['p1'], row['p2'], row['p3']]:
        player_freq[p] = player_freq.get(p, 0) + 1

freq_sorted = sorted(all_players, key=lambda p: -player_freq.get(p, 0))
pidx = {pid: idx for idx, pid in enumerate(freq_sorted)}
n_pl = len(pidx)

co = np.zeros((n_pl, n_pl), dtype=int)
for _, row in train_data.iterrows():
    players = [row['p1'], row['p2'], row['p3']]
    idxs = [pidx[p] for p in players if p in pidx]
    for i in range(len(idxs)):
        for j in range(i + 1, len(idxs)):
            co[idxs[i], idxs[j]] += 1
            co[idxs[j], idxs[i]] += 1

rcm = reverse_cuthill_mckee(csr_matrix(co))

# ============================================================
# Шаг 4: Бинарные матрицы
# ============================================================

I_trn = np.zeros((len(train_agg), n_pl), dtype=int)
y_trn = np.zeros(len(train_agg), dtype=float)
for i, (_, row) in enumerate(train_agg.iterrows()):
    for p in [row['p1'], row['p2'], row['p3']]:
        I_trn[i, pidx[p]] = 1
    y_trn[i] = row['y']
I_trn = I_trn[:, rcm]

test_known = test_agg[
    test_agg['p1'].isin(pidx) &
    test_agg['p2'].isin(pidx) &
    test_agg['p3'].isin(pidx)
].reset_index(drop=True)
test_known['in_train'] = test_known['lineId'].isin(set(train_agg['lineId']))

I_tst = np.zeros((len(test_known), n_pl), dtype=int)
y_tst = np.zeros(len(test_known), dtype=float)
for i, (_, row) in enumerate(test_known.iterrows()):
    for p in [row['p1'], row['p2'], row['p3']]:
        I_tst[i, pidx[p]] = 1
    y_tst[i] = row['y']
I_tst = I_tst[:, rcm]

good = [k for k in range(I_trn.shape[1]) if I_trn[:, k].max() == 1]
bad = [k for k in range(I_trn.shape[1]) if I_trn[:, k].max() == 0]

if len(bad) > 0:
    mask = np.ones(len(I_tst), dtype=bool)
    for k in bad:
        mask &= (I_tst[:, k] == 0)
    I_tst = I_tst[mask]
    y_tst = y_tst[mask]
    test_known = test_known[mask].reset_index(drop=True)

I_trn = I_trn[:, good]
I_tst = I_tst[:, good]

majority_acc = max(y_tst.mean(), 1 - y_tst.mean())

print(f"\n  I_trn: {I_trn.shape}, I_tst: {I_tst.shape}")
print(f"  Majority: {majority_acc:.3f}")
print(f"  Новых троек: {(~test_known['in_train']).sum()} / {len(test_known)}")

# ============================================================
# Шаг 5: ANOVA (без ALS) — baseline
# ============================================================

print(f"\n{'='*65}")
print(f"ANOVA order=1 (без ALS):")
print(f"{'='*65}")

for r in [2, 3, 5]:
    Y0 = teneva.anova(I_trn, y_trn, r=r, order=1)
    y_p_trn = teneva.get_many(Y0, I_trn)
    y_p_tst = teneva.get_many(Y0, I_tst)
    trn_acc = np.mean((y_p_trn > 0.5) == y_trn)
    tst_acc = np.mean((y_p_tst > 0.5) == y_tst)
    print(f"  r={r}: Train Acc={trn_acc:.4f}, Test Acc={tst_acc:.4f}")

# ============================================================
# Шаг 6: ANOVA + ALS с регуляризацией
# ============================================================

ranks = [2, 3, 5]
lambs = [0.01, 0.1, 1.0, 10.0]

results = {}

for r in ranks:
    for lamb in lambs:
        key = (r, lamb)
        print(f"\n{'='*65}")
        print(f"r={r}, λ={lamb}")
        print(f"{'='*65}")

        Y0 = teneva.anova(I_trn, y_trn, r=r, order=1)

        t = tpc()
        Y = teneva.als(
            I_trn, y_trn, Y0,
            nswp=30,
            lamb=lamb,
            I_vld=I_tst, y_vld=y_tst,
            e_vld=1e-3,
            log=True
        )
        t_als = tpc() - t

        y_p_trn = teneva.get_many(Y, I_trn)
        y_p_tst = teneva.get_many(Y, I_tst)

        trn_acc = np.mean((y_p_trn > 0.5) == y_trn)
        tst_acc = np.mean((y_p_tst > 0.5) == y_tst)
        trn_mae = np.mean(np.abs(y_p_trn - y_trn))
        tst_mae = np.mean(np.abs(y_p_tst - y_tst))

        try:
            tst_auc = roc_auc_score(y_tst, y_p_tst)
        except:
            tst_auc = 0.0

        known_mask = test_known['in_train'].values
        new_mask = ~known_mask
        acc_known = np.mean((y_p_tst[known_mask] > 0.5) == y_tst[known_mask]) if known_mask.sum() > 0 else 0
        acc_new = np.mean((y_p_tst[new_mask] > 0.5) == y_tst[new_mask]) if new_mask.sum() > 0 else 0

        print(f"  Train Acc={trn_acc:.4f}, Test Acc={tst_acc:.4f}, "
              f"AUC={tst_auc:.4f}, MAE={tst_mae:.4f} ({t_als:.1f}s)")
        print(f"  Знакомые={acc_known:.4f} (n={known_mask.sum()}), "
              f"Новые={acc_new:.4f} (n={new_mask.sum()})")

        results[key] = {
            'r': r, 'lamb': lamb,
            'trn_acc': trn_acc, 'tst_acc': tst_acc,
            'tst_auc': tst_auc,
            'trn_mae': trn_mae, 'tst_mae': tst_mae,
            'acc_known': acc_known, 'acc_new': acc_new,
            'y_pred': y_p_tst, 'time': t_als,
        }

# ============================================================
# Шаг 7: Итоговая таблица
# ============================================================

print(f"\n{'='*75}")
print(f"ИТОГОВАЯ ТАБЛИЦА:")
print(f"{'='*75}")
print(f"{'r':>3} {'λ':>8} {'Trn Acc':>8} {'Tst Acc':>8} "
      f"{'AUC':>8} {'Tst MAE':>8} {'Знаком':>8} {'Новые':>8} {'Time':>6}")
print(f"{'-'*75}")
print(f"{'':>3} {'Maj':>8} {'':>8} {majority_acc:>8.4f}")

for key in sorted(results.keys()):
    res = results[key]
    print(f"{res['r']:>3} {res['lamb']:>8.2f} {res['trn_acc']:>8.4f} "
          f"{res['tst_acc']:>8.4f} {res['tst_auc']:>8.4f} "
          f"{res['tst_mae']:>8.4f} {res['acc_known']:>8.4f} "
          f"{res['acc_new']:>8.4f} {res['time']:>5.1f}s")

best_key = max(results, key=lambda k: results[k]['tst_acc'])
best = results[best_key]
print(f"\n  ★ Лучшая: r={best['r']}, λ={best['lamb']}, "
      f"Test Acc={best['tst_acc']:.4f}, AUC={best['tst_auc']:.4f}")

# ============================================================
# Шаг 8: Визуализация
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1: Test Acc vs λ
for r in ranks:
    accs = [results[(r, l)]['tst_acc'] for l in lambs]
    axes[0,0].plot(range(len(lambs)), accs, 'o-', label=f'r={r}', linewidth=2)
axes[0,0].axhline(majority_acc, color='red', linestyle='--',
                   label=f'Majority={majority_acc:.3f}')
axes[0,0].set_xticks(range(len(lambs)))
axes[0,0].set_xticklabels([str(l) for l in lambs])
axes[0,0].set_xlabel('λ')
axes[0,0].set_ylabel('Test Accuracy')
axes[0,0].set_title('Test Accuracy vs регуляризация')
axes[0,0].legend()

# 2: Переобучение: Train - Test gap
for r in ranks:
    gaps = [results[(r, l)]['trn_acc'] - results[(r, l)]['tst_acc'] for l in lambs]
    axes[0,1].plot(range(len(lambs)), gaps, 'o-', label=f'r={r}', linewidth=2)
axes[0,1].axhline(0, color='gray', linestyle=':')
axes[0,1].set_xticks(range(len(lambs)))
axes[0,1].set_xticklabels([str(l) for l in lambs])
axes[0,1].set_xlabel('λ')
axes[0,1].set_ylabel('Train Acc - Test Acc')
axes[0,1].set_title('Разрыв Train-Test (переобучение)')
axes[0,1].legend()

# 3: AUC vs λ
for r in ranks:
    aucs = [results[(r, l)]['tst_auc'] for l in lambs]
    axes[1,0].plot(range(len(lambs)), aucs, 'o-', label=f'r={r}', linewidth=2)
axes[1,0].axhline(0.5, color='red', linestyle='--')
axes[1,0].set_xticks(range(len(lambs)))
axes[1,0].set_xticklabels([str(l) for l in lambs])
axes[1,0].set_xlabel('λ')
axes[1,0].set_ylabel('Test AUC')
axes[1,0].set_title('AUC vs регуляризация')
axes[1,0].legend()

# 4: Лучшая модель — предсказания
y_best = best['y_pred']
cf_actual = test_known['cf_pct'].values
colors = ['steelblue' if t else 'coral' for t in test_known['in_train']]
axes[1,1].scatter(cf_actual, y_best, alpha=0.5, s=25, c=colors)
axes[1,1].axhline(0.5, color='gray', linestyle=':')
axes[1,1].axvline(0.5, color='gray', linestyle=':')
axes[1,1].set_xlabel('Actual CF%')
axes[1,1].set_ylabel('Predicted')
axes[1,1].set_title(f'Лучшая модель (r={best["r"]}, λ={best["lamb"]})\n'
                     f'Синие=знакомые, Красные=новые')

plt.suptitle('TT-ANOVA + ALS с регуляризацией: сезон 2023',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('fig_tt_regularization.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import pandas as pd
import teneva
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import reverse_cuthill_mckee
from sklearn.metrics import accuracy_score, roc_auc_score, mean_squared_error
import matplotlib.pyplot as plt
from time import perf_counter as tpc

# ============================================================
# Шаг 1: Загрузка и split по дате (75/25)
# ============================================================

df = pd.read_csv('season_2023_filtered.csv')
df['gameDate'] = pd.to_datetime(df['gameDate'])

all_dates = sorted(df['gameDate'].unique())
split_date = all_dates[int(len(all_dates) * 0.75)]

train_data = df[df['gameDate'] <= split_date].copy()
test_data = df[df['gameDate'] > split_date].copy()

print(f"{'='*65}")
print(f"SPLIT ПО ДАТЕ:")
print(f"{'='*65}")
print(f"  Train: {len(train_data):,} записей, до {split_date.date()}")
print(f"  Test:  {len(test_data):,} записей")

# ============================================================
# Шаг 2: Агрегация по тройкам
# ============================================================

train_agg = train_data.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
).reset_index()
train_agg = train_agg[train_agg['n_games'] >= 3].reset_index(drop=True)
train_agg['cf_pct'] = train_agg['CF'] / (train_agg['CF'] + train_agg['CA'])

test_agg = test_data.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
).reset_index()
test_agg = test_agg[test_agg['n_games'] >= 3].reset_index(drop=True)
test_agg['cf_pct'] = test_agg['CF'] / (test_agg['CF'] + test_agg['CA'])

print(f"  Train троек: {len(train_agg)}, "
      f"CF% mean={train_agg['cf_pct'].mean():.3f}, "
      f"std={train_agg['cf_pct'].std():.3f}")
print(f"  Test троек:  {len(test_agg)}, "
      f"CF% mean={test_agg['cf_pct'].mean():.3f}")

# ============================================================
# Шаг 3: Индексация + RCM
# ============================================================

all_players = set()
for _, row in train_agg.iterrows():
    all_players.update([row['p1'], row['p2'], row['p3']])

player_freq = {}
for _, row in train_data.iterrows():
    for p in [row['p1'], row['p2'], row['p3']]:
        player_freq[p] = player_freq.get(p, 0) + 1

freq_sorted = sorted(all_players, key=lambda p: -player_freq.get(p, 0))
pidx = {pid: idx for idx, pid in enumerate(freq_sorted)}
n_pl = len(pidx)

co = np.zeros((n_pl, n_pl), dtype=int)
for _, row in train_data.iterrows():
    players = [row['p1'], row['p2'], row['p3']]
    idxs = [pidx[p] for p in players if p in pidx]
    for i in range(len(idxs)):
        for j in range(i + 1, len(idxs)):
            co[idxs[i], idxs[j]] += 1
            co[idxs[j], idxs[i]] += 1

rcm = reverse_cuthill_mckee(csr_matrix(co))

# ============================================================
# Шаг 4: Бинарные матрицы — ТАРГЕТ = CF%
# ============================================================

# Train: таргет = CF% (непрерывный!)
I_trn = np.zeros((len(train_agg), n_pl), dtype=int)
y_trn = np.zeros(len(train_agg), dtype=float)
for i, (_, row) in enumerate(train_agg.iterrows()):
    for p in [row['p1'], row['p2'], row['p3']]:
        I_trn[i, pidx[p]] = 1
    y_trn[i] = row['cf_pct']  # <-- CF%, не бинарный!
I_trn = I_trn[:, rcm]

# Test
test_known = test_agg[
    test_agg['p1'].isin(pidx) &
    test_agg['p2'].isin(pidx) &
    test_agg['p3'].isin(pidx)
].reset_index(drop=True)
test_known['in_train'] = test_known['lineId'].isin(set(train_agg['lineId']))

I_tst = np.zeros((len(test_known), n_pl), dtype=int)
y_tst_cf = np.zeros(len(test_known), dtype=float)
for i, (_, row) in enumerate(test_known.iterrows()):
    for p in [row['p1'], row['p2'], row['p3']]:
        I_tst[i, pidx[p]] = 1
    y_tst_cf[i] = row['cf_pct']  # <-- CF%
I_tst = I_tst[:, rcm]

y_tst_bin = (y_tst_cf > 0.5).astype(float)  # для оценки accuracy

# Убираем мёртвые столбцы
good = [k for k in range(I_trn.shape[1]) if I_trn[:, k].max() == 1]
bad = [k for k in range(I_trn.shape[1]) if I_trn[:, k].max() == 0]

if len(bad) > 0:
    mask = np.ones(len(I_tst), dtype=bool)
    for k in bad:
        mask &= (I_tst[:, k] == 0)
    I_tst = I_tst[mask]
    y_tst_cf = y_tst_cf[mask]
    y_tst_bin = y_tst_bin[mask]
    test_known = test_known[mask].reset_index(drop=True)

I_trn = I_trn[:, good]
I_tst = I_tst[:, good]

majority_acc = max(y_tst_bin.mean(), 1 - y_tst_bin.mean())

print(f"\n{'='*65}")
print(f"ДАННЫЕ (таргет = CF%):")
print(f"{'='*65}")
print(f"  I_trn: {I_trn.shape}")
print(f"  y_trn: CF% range [{y_trn.min():.3f}, {y_trn.max():.3f}], "
      f"mean={y_trn.mean():.3f}")
print(f"  I_tst: {I_tst.shape}")
print(f"  y_tst: CF% mean={y_tst_cf.mean():.3f}, "
      f"binary y=1: {y_tst_bin.mean():.3f}")
print(f"  Majority: {majority_acc:.3f}")
print(f"  Новых в test: {(~test_known['in_train']).sum()} / {len(test_known)}")

# ============================================================
# Шаг 5: ANOVA + ALS
# ============================================================

ranks = [2, 3, 5]
lambs = [0.01, 0.1, 1.0, 10.0]

results = {}

# ANOVA baseline
print(f"\n{'='*65}")
print(f"ANOVA order=1 (без ALS):")
print(f"{'='*65}")

for r in ranks:
    Y0 = teneva.anova(I_trn, y_trn, r=r, order=1)
    y_p = teneva.get_many(Y0, I_tst)
    y_p = np.clip(y_p, 0, 1)
    acc = np.mean((y_p > 0.5) == y_tst_bin)
    mse = np.mean((y_p - y_tst_cf) ** 2)
    print(f"  r={r}: Test Acc={acc:.4f}, MSE={mse:.4f}")

# ANOVA + ALS
for r in ranks:
    for lamb in lambs:
        key = (r, lamb)
        print(f"\n{'='*65}")
        print(f"r={r}, λ={lamb}")
        print(f"{'='*65}")

        Y0 = teneva.anova(I_trn, y_trn, r=r, order=1)

        t = tpc()
        Y = teneva.als(
            I_trn, y_trn, Y0,
            nswp=30,
            lamb=lamb,
            I_vld=I_tst, y_vld=y_tst_cf,  # <-- валидируем тоже по CF%
            e_vld=1e-3,
            log=True
        )
        t_als = tpc() - t

        y_p_trn = teneva.get_many(Y, I_trn)
        y_p_tst = teneva.get_many(Y, I_tst)

        y_p_trn_clip = np.clip(y_p_trn, 0, 1)
        y_p_tst_clip = np.clip(y_p_tst, 0, 1)

        # Регрессионные метрики
        trn_mse = np.mean((y_p_trn_clip - y_trn) ** 2)
        tst_mse = np.mean((y_p_tst_clip - y_tst_cf) ** 2)
        trn_mae = np.mean(np.abs(y_p_trn_clip - y_trn))
        tst_mae = np.mean(np.abs(y_p_tst_clip - y_tst_cf))

        # Классификационные метрики (порог 0.5)
        trn_acc = np.mean((y_p_trn_clip > 0.5) == (y_trn > 0.5))
        tst_acc = np.mean((y_p_tst_clip > 0.5) == y_tst_bin)

        try:
            tst_auc = roc_auc_score(y_tst_bin, y_p_tst_clip)
        except:
            tst_auc = 0.0

        # Знакомые vs новые
        known_mask = test_known['in_train'].values
        new_mask = ~known_mask
        acc_known = np.mean((y_p_tst_clip[known_mask] > 0.5) == y_tst_bin[known_mask]) if known_mask.sum() > 0 else 0
        acc_new = np.mean((y_p_tst_clip[new_mask] > 0.5) == y_tst_bin[new_mask]) if new_mask.sum() > 0 else 0

        print(f"  Train: Acc={trn_acc:.4f}, MSE={trn_mse:.4f}, MAE={trn_mae:.4f}")
        print(f"  Test:  Acc={tst_acc:.4f}, MSE={tst_mse:.4f}, MAE={tst_mae:.4f}")
        print(f"  Test:  AUC={tst_auc:.4f}")
        print(f"  Знакомые={acc_known:.4f} (n={known_mask.sum()}), "
              f"Новые={acc_new:.4f} (n={new_mask.sum()})")
        print(f"  Pred range: [{y_p_tst.min():.3f}, {y_p_tst.max():.3f}]")

        results[key] = {
            'r': r, 'lamb': lamb,
            'trn_acc': trn_acc, 'tst_acc': tst_acc,
            'tst_auc': tst_auc,
            'trn_mse': trn_mse, 'tst_mse': tst_mse,
            'trn_mae': trn_mae, 'tst_mae': tst_mae,
            'acc_known': acc_known, 'acc_new': acc_new,
            'y_pred': y_p_tst_clip, 'time': t_als,
        }

# ============================================================
# Шаг 6: Итоговая таблица
# ============================================================

print(f"\n{'='*80}")
print(f"ИТОГОВАЯ ТАБЛИЦА (таргет = CF%):")
print(f"{'='*80}")
print(f"{'r':>3} {'λ':>7} {'Trn Acc':>8} {'Tst Acc':>8} {'AUC':>7} "
      f"{'Tst MSE':>8} {'Tst MAE':>8} {'Знак':>7} {'Нов':>7}")
print(f"{'-'*80}")
print(f"{'':>3} {'Maj':>7} {'':>8} {majority_acc:>8.3f}")

for key in sorted(results.keys()):
    res = results[key]
    print(f"{res['r']:>3} {res['lamb']:>7.2f} {res['trn_acc']:>8.4f} "
          f"{res['tst_acc']:>8.4f} {res['tst_auc']:>7.4f} "
          f"{res['tst_mse']:>8.4f} {res['tst_mae']:>8.4f} "
          f"{res['acc_known']:>7.4f} {res['acc_new']:>7.4f}")

best_key = max(results, key=lambda k: results[k]['tst_acc'])
best = results[best_key]
print(f"\n  ★ Лучшая: r={best['r']}, λ={best['lamb']}, "
      f"Test Acc={best['tst_acc']:.4f}, AUC={best['tst_auc']:.4f}")

# Сравнение с бинарным таргетом
print(f"\n  Ранее (бинарный y): лучшая Test Acc=0.6275, AUC=0.6268")
print(f"  Сейчас (CF%):       лучшая Test Acc={best['tst_acc']:.4f}, "
      f"AUC={best['tst_auc']:.4f}")

# ============================================================
# Шаг 7: Визуализация
# ============================================================

fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 1: Test Acc vs λ
for r in ranks:
    accs = [results[(r, l)]['tst_acc'] for l in lambs]
    axes[0,0].plot(range(len(lambs)), accs, 'o-', label=f'r={r}', linewidth=2)
axes[0,0].axhline(majority_acc, color='red', linestyle='--',
                   label=f'Majority={majority_acc:.3f}')
axes[0,0].set_xticks(range(len(lambs)))
axes[0,0].set_xticklabels([str(l) for l in lambs])
axes[0,0].set_xlabel('λ')
axes[0,0].set_ylabel('Test Accuracy')
axes[0,0].set_title('Test Accuracy vs λ')
axes[0,0].legend()

# 2: Train-Test gap
for r in ranks:
    gaps = [results[(r, l)]['trn_acc'] - results[(r, l)]['tst_acc'] for l in lambs]
    axes[0,1].plot(range(len(lambs)), gaps, 'o-', label=f'r={r}', linewidth=2)
axes[0,1].axhline(0, color='gray', linestyle=':')
axes[0,1].set_xticks(range(len(lambs)))
axes[0,1].set_xticklabels([str(l) for l in lambs])
axes[0,1].set_xlabel('λ')
axes[0,1].set_ylabel('Train - Test Acc')
axes[0,1].set_title('Переобучение')
axes[0,1].legend()

# 3: AUC
for r in ranks:
    aucs = [results[(r, l)]['tst_auc'] for l in lambs]
    axes[0,2].plot(range(len(lambs)), aucs, 'o-', label=f'r={r}', linewidth=2)
axes[0,2].axhline(0.5, color='red', linestyle='--')
axes[0,2].set_xticks(range(len(lambs)))
axes[0,2].set_xticklabels([str(l) for l in lambs])
axes[0,2].set_xlabel('λ')
axes[0,2].set_ylabel('Test AUC')
axes[0,2].set_title('AUC vs λ')
axes[0,2].legend()

# 4: Test MSE
for r in ranks:
    mses = [results[(r, l)]['tst_mse'] for l in lambs]
    axes[1,0].plot(range(len(lambs)), mses, 'o-', label=f'r={r}', linewidth=2)
axes[1,0].set_xticks(range(len(lambs)))
axes[1,0].set_xticklabels([str(l) for l in lambs])
axes[1,0].set_xlabel('λ')
axes[1,0].set_ylabel('Test MSE')
axes[1,0].set_title('MSE vs λ')
axes[1,0].legend()

# 5: Predicted CF% vs Actual CF%
y_best = best['y_pred']
cf_actual = test_known['cf_pct'].values
colors = ['steelblue' if t else 'coral' for t in test_known['in_train']]
axes[1,1].scatter(cf_actual, y_best, alpha=0.5, s=25, c=colors)
axes[1,1].plot([0.3, 0.7], [0.3, 0.7], 'r--')
axes[1,1].axhline(0.5, color='gray', linestyle=':')
axes[1,1].axvline(0.5, color='gray', linestyle=':')
axes[1,1].set_xlabel('Actual CF%')
axes[1,1].set_ylabel('Predicted CF%')
axes[1,1].set_title(f'Лучшая (r={best["r"]}, λ={best["lamb"]})')

# 6: Распределение предсказаний
axes[1,2].hist(y_best[y_tst_bin==1], bins=20, alpha=0.6,
               color='steelblue', label='CF%>50%', density=True)
axes[1,2].hist(y_best[y_tst_bin==0], bins=20, alpha=0.6,
               color='coral', label='CF%≤50%', density=True)
axes[1,2].axvline(0.5, color='black', linestyle='--')
axes[1,2].set_xlabel('Predicted CF%')
axes[1,2].set_title('Распределение предсказаний')
axes[1,2].legend()

plt.suptitle('TT-ANOVA + ALS: таргет = CF% (сезон 2023)',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('fig_tt_cf_target.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import pandas as pd
import teneva
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import reverse_cuthill_mckee
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.linear_model import Ridge, LogisticRegression
import matplotlib.pyplot as plt
from time import perf_counter as tpc

# ============================================================
# Шаг 1: Загрузка и split
# ============================================================

df = pd.read_csv('season_2023_filtered.csv')
df['gameDate'] = pd.to_datetime(df['gameDate'])

all_dates = sorted(df['gameDate'].unique())
split_date = all_dates[int(len(all_dates) * 0.75)]

train_data = df[df['gameDate'] <= split_date].copy()
test_data = df[df['gameDate'] > split_date].copy()

# ============================================================
# Шаг 2: Агрегация по тройкам
# ============================================================

train_agg = train_data.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
).reset_index()
train_agg = train_agg[train_agg['n_games'] >= 3].reset_index(drop=True)
train_agg['cf_pct'] = train_agg['CF'] / (train_agg['CF'] + train_agg['CA'])

test_agg = test_data.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
).reset_index()
test_agg = test_agg[test_agg['n_games'] >= 3].reset_index(drop=True)
test_agg['cf_pct'] = test_agg['CF'] / (test_agg['CF'] + test_agg['CA'])

# ============================================================
# Шаг 3: Индексация + RCM + бинарные матрицы
# ============================================================

all_players = set()
for _, row in train_agg.iterrows():
    all_players.update([row['p1'], row['p2'], row['p3']])

player_freq = {}
for _, row in train_data.iterrows():
    for p in [row['p1'], row['p2'], row['p3']]:
        player_freq[p] = player_freq.get(p, 0) + 1

freq_sorted = sorted(all_players, key=lambda p: -player_freq.get(p, 0))
pidx = {pid: idx for idx, pid in enumerate(freq_sorted)}
n_pl = len(pidx)

co = np.zeros((n_pl, n_pl), dtype=int)
for _, row in train_data.iterrows():
    players = [row['p1'], row['p2'], row['p3']]
    idxs = [pidx[p] for p in players if p in pidx]
    for i in range(len(idxs)):
        for j in range(i + 1, len(idxs)):
            co[idxs[i], idxs[j]] += 1
            co[idxs[j], idxs[i]] += 1

rcm = reverse_cuthill_mckee(csr_matrix(co))

I_trn = np.zeros((len(train_agg), n_pl), dtype=int)
y_trn = np.zeros(len(train_agg), dtype=float)
for i, (_, row) in enumerate(train_agg.iterrows()):
    for p in [row['p1'], row['p2'], row['p3']]:
        I_trn[i, pidx[p]] = 1
    y_trn[i] = row['cf_pct']
I_trn = I_trn[:, rcm]

test_known = test_agg[
    test_agg['p1'].isin(pidx) &
    test_agg['p2'].isin(pidx) &
    test_agg['p3'].isin(pidx)
].reset_index(drop=True)
test_known['in_train'] = test_known['lineId'].isin(set(train_agg['lineId']))

I_tst = np.zeros((len(test_known), n_pl), dtype=int)
y_tst_cf = np.zeros(len(test_known), dtype=float)
for i, (_, row) in enumerate(test_known.iterrows()):
    for p in [row['p1'], row['p2'], row['p3']]:
        I_tst[i, pidx[p]] = 1
    y_tst_cf[i] = row['cf_pct']
I_tst = I_tst[:, rcm]

y_tst_bin = (y_tst_cf > 0.5).astype(float)

good = [k for k in range(I_trn.shape[1]) if I_trn[:, k].max() == 1]
bad = [k for k in range(I_trn.shape[1]) if I_trn[:, k].max() == 0]
if len(bad) > 0:
    mask = np.ones(len(I_tst), dtype=bool)
    for k in bad:
        mask &= (I_tst[:, k] == 0)
    I_tst = I_tst[mask]
    y_tst_cf = y_tst_cf[mask]
    y_tst_bin = y_tst_bin[mask]
    test_known = test_known[mask].reset_index(drop=True)
I_trn = I_trn[:, good]
I_tst = I_tst[:, good]

known_mask = test_known['in_train'].values
new_mask = ~known_mask
majority_acc = max(y_tst_bin.mean(), 1 - y_tst_bin.mean())

print(f"{'='*65}")
print(f"ДАННЫЕ:")
print(f"{'='*65}")
print(f"  Train: {I_trn.shape[0]} троек, {I_trn.shape[1]} игроков")
print(f"  Test:  {I_tst.shape[0]} троек")
print(f"  Majority: {majority_acc:.3f}")
print(f"  Знакомых: {known_mask.sum()}, Новых: {new_mask.sum()}")

# ============================================================
# Шаг 4: Бейзлайн 1 — среднее CF% по train
# ============================================================

print(f"\n{'='*65}")
print(f"БЕЙЗЛАЙН 1: Среднее CF% по train")
print(f"{'='*65}")

mean_cf = y_trn.mean()
pred_mean = np.full(len(y_tst_cf), mean_cf)

acc_mean = np.mean((pred_mean > 0.5) == y_tst_bin)
mse_mean = np.mean((pred_mean - y_tst_cf) ** 2)
mae_mean = np.mean(np.abs(pred_mean - y_tst_cf))

print(f"  Предсказание для всех: {mean_cf:.4f}")
print(f"  Test Acc: {acc_mean:.4f}")
print(f"  Test MSE: {mse_mean:.4f}, MAE: {mae_mean:.4f}")
# AUC невозможен — все предсказания одинаковые

# ============================================================
# Шаг 5: Бейзлайн 2 — среднее CF% трёх игроков
# ============================================================

print(f"\n{'='*65}")
print(f"БЕЙЗЛАЙН 2: Среднее индивидуальных CF% трёх игроков")
print(f"{'='*65}")

# Считаем CF% каждого игрока из train
player_cf = {}
for _, row in train_data.iterrows():
    for p in [row['p1'], row['p2'], row['p3']]:
        if p not in player_cf:
            player_cf[p] = {'CF': 0, 'CA': 0}
        player_cf[p]['CF'] += row['CF']
        player_cf[p]['CA'] += row['CA']

player_cf_pct = {}
for p, stats in player_cf.items():
    total = stats['CF'] + stats['CA']
    player_cf_pct[p] = stats['CF'] / total if total > 0 else 0.5

# Статистики по игрокам
pvals = list(player_cf_pct.values())
print(f"  Игроков с CF%: {len(pvals)}")
print(f"  Индивид. CF%: mean={np.mean(pvals):.3f}, "
      f"std={np.std(pvals):.3f}, "
      f"range=[{np.min(pvals):.3f}, {np.max(pvals):.3f}]")

# Предсказание = среднее CF% трёх игроков
pred_avg = np.zeros(len(test_known))
for i, (_, row) in enumerate(test_known.iterrows()):
    cf_vals = [player_cf_pct.get(row[p], 0.5) for p in ['p1', 'p2', 'p3']]
    pred_avg[i] = np.mean(cf_vals)

acc_avg = np.mean((pred_avg > 0.5) == y_tst_bin)
mse_avg = np.mean((pred_avg - y_tst_cf) ** 2)
mae_avg = np.mean(np.abs(pred_avg - y_tst_cf))

try:
    auc_avg = roc_auc_score(y_tst_bin, pred_avg)
except:
    auc_avg = 0.0

acc_avg_known = np.mean((pred_avg[known_mask] > 0.5) == y_tst_bin[known_mask]) if known_mask.sum() > 0 else 0
acc_avg_new = np.mean((pred_avg[new_mask] > 0.5) == y_tst_bin[new_mask]) if new_mask.sum() > 0 else 0

print(f"\n  Test Acc: {acc_avg:.4f}, AUC: {auc_avg:.4f}")
print(f"  Test MSE: {mse_avg:.4f}, MAE: {mae_avg:.4f}")
print(f"  Знакомые: {acc_avg_known:.4f}, Новые: {acc_avg_new:.4f}")

# ============================================================
# Шаг 6: Бейзлайн 3 — Ridge Regression
# ============================================================

print(f"\n{'='*65}")
print(f"БЕЙЗЛАЙН 3: Ridge Regression (на бинарной матрице)")
print(f"{'='*65}")

best_ridge_acc = 0
best_ridge_alpha = None

for alpha in [0.01, 0.1, 1.0, 10.0, 100.0]:
    ridge = Ridge(alpha=alpha)
    ridge.fit(I_trn, y_trn)
    pred_ridge = ridge.predict(I_tst)
    pred_ridge = np.clip(pred_ridge, 0, 1)
    acc_r = np.mean((pred_ridge > 0.5) == y_tst_bin)
    
    if acc_r > best_ridge_acc:
        best_ridge_acc = acc_r
        best_ridge_alpha = alpha
        best_ridge_pred = pred_ridge.copy()

ridge = Ridge(alpha=best_ridge_alpha)
ridge.fit(I_trn, y_trn)
pred_ridge = np.clip(ridge.predict(I_tst), 0, 1)
pred_ridge_trn = np.clip(ridge.predict(I_trn), 0, 1)

acc_ridge = np.mean((pred_ridge > 0.5) == y_tst_bin)
acc_ridge_trn = np.mean((pred_ridge_trn > 0.5) == (y_trn > 0.5))
mse_ridge = np.mean((pred_ridge - y_tst_cf) ** 2)
mae_ridge = np.mean(np.abs(pred_ridge - y_tst_cf))

try:
    auc_ridge = roc_auc_score(y_tst_bin, pred_ridge)
except:
    auc_ridge = 0.0

acc_ridge_known = np.mean((pred_ridge[known_mask] > 0.5) == y_tst_bin[known_mask]) if known_mask.sum() > 0 else 0
acc_ridge_new = np.mean((pred_ridge[new_mask] > 0.5) == y_tst_bin[new_mask]) if new_mask.sum() > 0 else 0

print(f"  Лучший alpha: {best_ridge_alpha}")
print(f"  Train Acc: {acc_ridge_trn:.4f}")
print(f"  Test Acc:  {acc_ridge:.4f}, AUC: {auc_ridge:.4f}")
print(f"  Test MSE:  {mse_ridge:.4f}, MAE: {mae_ridge:.4f}")
print(f"  Знакомые: {acc_ridge_known:.4f}, Новые: {acc_ridge_new:.4f}")

# ============================================================
# Шаг 7: TT-ANOVA + ALS (лучшая конфигурация)
# ============================================================

print(f"\n{'='*65}")
print(f"TT-ANOVA + ALS (r=2, λ=0.1)")
print(f"{'='*65}")

Y0 = teneva.anova(I_trn, y_trn, r=2, order=1)

Y = teneva.als(
    I_trn, y_trn, Y0,
    nswp=30, lamb=0.1,
    I_vld=I_tst, y_vld=y_tst_cf,
    e_vld=1e-3, log=True
)

pred_tt_trn = np.clip(teneva.get_many(Y, I_trn), 0, 1)
pred_tt = np.clip(teneva.get_many(Y, I_tst), 0, 1)

acc_tt = np.mean((pred_tt > 0.5) == y_tst_bin)
acc_tt_trn = np.mean((pred_tt_trn > 0.5) == (y_trn > 0.5))
mse_tt = np.mean((pred_tt - y_tst_cf) ** 2)
mae_tt = np.mean(np.abs(pred_tt - y_tst_cf))

try:
    auc_tt = roc_auc_score(y_tst_bin, pred_tt)
except:
    auc_tt = 0.0

acc_tt_known = np.mean((pred_tt[known_mask] > 0.5) == y_tst_bin[known_mask]) if known_mask.sum() > 0 else 0
acc_tt_new = np.mean((pred_tt[new_mask] > 0.5) == y_tst_bin[new_mask]) if new_mask.sum() > 0 else 0

print(f"  Train Acc: {acc_tt_trn:.4f}")
print(f"  Test Acc:  {acc_tt:.4f}, AUC: {auc_tt:.4f}")
print(f"  Test MSE:  {mse_tt:.4f}, MAE: {mae_tt:.4f}")
print(f"  Знакомые: {acc_tt_known:.4f}, Новые: {acc_tt_new:.4f}")

# ============================================================
# Шаг 8: Итоговая таблица
# ============================================================

print(f"\n{'='*75}")
print(f"СРАВНИТЕЛЬНАЯ ТАБЛИЦА:")
print(f"{'='*75}")
print(f"{'Модель':<30s} {'Tst Acc':>8} {'AUC':>8} {'MSE':>8} "
      f"{'MAE':>8} {'Знак':>7} {'Нов':>7}")
print(f"{'-'*75}")
print(f"{'Majority':<30s} {majority_acc:>8.4f} {'—':>8} {'—':>8} "
      f"{'—':>8} {'—':>7} {'—':>7}")
print(f"{'Среднее train CF%':<30s} {acc_mean:>8.4f} {'—':>8} {mse_mean:>8.4f} "
      f"{mae_mean:>8.4f} {'—':>7} {'—':>7}")
print(f"{'Avg CF% 3 игроков':<30s} {acc_avg:>8.4f} {auc_avg:>8.4f} {mse_avg:>8.4f} "
      f"{mae_avg:>8.4f} {acc_avg_known:>7.4f} {acc_avg_new:>7.4f}")
print(f"{'Ridge (α=' + str(best_ridge_alpha) + ')':<30s} {acc_ridge:>8.4f} {auc_ridge:>8.4f} {mse_ridge:>8.4f} "
      f"{mae_ridge:>8.4f} {acc_ridge_known:>7.4f} {acc_ridge_new:>7.4f}")
print(f"{'TT-ANOVA+ALS (r=2, λ=0.1)':<30s} {acc_tt:>8.4f} {auc_tt:>8.4f} {mse_tt:>8.4f} "
      f"{mae_tt:>8.4f} {acc_tt_known:>7.4f} {acc_tt_new:>7.4f}")

# ============================================================
# Шаг 9: Визуализация
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1: Accuracy сравнение
models = ['Majority', 'Mean CF%', 'Avg 3 игр.', 'Ridge', 'TT-ANOVA+ALS']
accs = [majority_acc, acc_mean, acc_avg, acc_ridge, acc_tt]
colors = ['gray', 'lightcoral', 'coral', 'steelblue', 'darkblue']
bars = axes[0,0].barh(models, accs, color=colors)
axes[0,0].set_xlabel('Test Accuracy')
axes[0,0].set_title('Сравнение моделей: Accuracy')
axes[0,0].set_xlim(0.4, 0.75)
for bar, val in zip(bars, accs):
    axes[0,0].text(val + 0.005, bar.get_y() + bar.get_height()/2,
                    f'{val:.3f}', va='center', fontweight='bold')

# 2: Predicted vs Actual для всех моделей
axes[0,1].scatter(y_tst_cf, pred_avg, alpha=0.4, s=20, label='Avg 3 игр.', color='coral')
axes[0,1].scatter(y_tst_cf, pred_ridge, alpha=0.4, s=20, label='Ridge', color='steelblue')
axes[0,1].scatter(y_tst_cf, pred_tt, alpha=0.4, s=20, label='TT', color='darkblue')
axes[0,1].plot([0.3, 0.7], [0.3, 0.7], 'r--')
axes[0,1].set_xlabel('Actual CF%')
axes[0,1].set_ylabel('Predicted CF%')
axes[0,1].set_title('Predicted vs Actual')
axes[0,1].legend()

# 3: AUC сравнение
models_auc = ['Avg 3 игр.', 'Ridge', 'TT-ANOVA+ALS']
aucs = [auc_avg, auc_ridge, auc_tt]
colors_auc = ['coral', 'steelblue', 'darkblue']
bars = axes[1,0].barh(models_auc, aucs, color=colors_auc)
axes[1,0].axvline(0.5, color='red', linestyle='--', label='Random')
axes[1,0].set_xlabel('Test AUC')
axes[1,0].set_title('Сравнение моделей: AUC')
axes[1,0].set_xlim(0.4, 0.8)
for bar, val in zip(bars, aucs):
    axes[1,0].text(val + 0.005, bar.get_y() + bar.get_height()/2,
                    f'{val:.3f}', va='center', fontweight='bold')
axes[1,0].legend()

# 4: Знакомые vs Новые
x = np.arange(3)
w = 0.35
known_accs = [acc_avg_known, acc_ridge_known, acc_tt_known]
new_accs = [acc_avg_new, acc_ridge_new, acc_tt_new]
axes[1,1].bar(x - w/2, known_accs, w, label='Знакомые', color='steelblue')
axes[1,1].bar(x + w/2, new_accs, w, label='Новые', color='coral')
axes[1,1].set_xticks(x)
axes[1,1].set_xticklabels(['Avg 3 игр.', 'Ridge', 'TT'])
axes[1,1].axhline(majority_acc, color='red', linestyle='--', label='Majority')
axes[1,1].set_ylabel('Accuracy')
axes[1,1].set_title('Знакомые vs Новые тройки')
axes[1,1].legend()

plt.suptitle('Сравнение моделей: сезон 2023',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('fig_model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

# Как строится Ridge

У нас бинарная матрица I (532 × 470):
  строки = тройки
  столбцы = игроки
  значение = 1 если игрок в тройке

Модель: CF%_тройки = w₁·x₁ + w₂·x₂ + ... + w₄₇₀·x₄₇₀ + b

Где xᵢ ∈ {0,1} — участвует ли i-й игрок.

Так как в каждой тройке ровно 3 единицы:
  CF%_тройки = wₐ + w_b + w_c + b

  где a, b, c — три игрока в тройке.

По сути: каждый игрок получает "вес" wᵢ,
и CF% тройки = сумма весов трёх игроков + сдвиг.

# ANOVA order = 2

```python
import numpy as np
import pandas as pd
import teneva
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import reverse_cuthill_mckee
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
from time import perf_counter as tpc

# ============================================================
# Подготовка данных (как раньше)
# ============================================================

df = pd.read_csv('season_2023_filtered.csv')
df['gameDate'] = pd.to_datetime(df['gameDate'])

all_dates = sorted(df['gameDate'].unique())
split_date = all_dates[int(len(all_dates) * 0.75)]

train_data = df[df['gameDate'] <= split_date].copy()
test_data = df[df['gameDate'] > split_date].copy()

train_agg = train_data.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
).reset_index()
train_agg = train_agg[train_agg['n_games'] >= 3].reset_index(drop=True)
train_agg['cf_pct'] = train_agg['CF'] / (train_agg['CF'] + train_agg['CA'])

test_agg = test_data.groupby('lineId').agg(
    CF=('CF', 'sum'), CA=('CA', 'sum'),
    n_games=('gameId', 'nunique'),
    p1=('p1', 'first'), p2=('p2', 'first'), p3=('p3', 'first'),
    name=('name', 'first'),
).reset_index()
test_agg = test_agg[test_agg['n_games'] >= 3].reset_index(drop=True)
test_agg['cf_pct'] = test_agg['CF'] / (test_agg['CF'] + test_agg['CA'])

all_players = set()
for _, row in train_agg.iterrows():
    all_players.update([row['p1'], row['p2'], row['p3']])

player_freq = {}
for _, row in train_data.iterrows():
    for p in [row['p1'], row['p2'], row['p3']]:
        player_freq[p] = player_freq.get(p, 0) + 1

freq_sorted = sorted(all_players, key=lambda p: -player_freq.get(p, 0))
pidx = {pid: idx for idx, pid in enumerate(freq_sorted)}
n_pl = len(pidx)

co = np.zeros((n_pl, n_pl), dtype=int)
for _, row in train_data.iterrows():
    players = [row['p1'], row['p2'], row['p3']]
    idxs = [pidx[p] for p in players if p in pidx]
    for i in range(len(idxs)):
        for j in range(i + 1, len(idxs)):
            co[idxs[i], idxs[j]] += 1
            co[idxs[j], idxs[i]] += 1

rcm = reverse_cuthill_mckee(csr_matrix(co))

I_trn = np.zeros((len(train_agg), n_pl), dtype=int)
y_trn = np.zeros(len(train_agg), dtype=float)
for i, (_, row) in enumerate(train_agg.iterrows()):
    for p in [row['p1'], row['p2'], row['p3']]:
        I_trn[i, pidx[p]] = 1
    y_trn[i] = row['cf_pct']
I_trn = I_trn[:, rcm]

test_known = test_agg[
    test_agg['p1'].isin(pidx) &
    test_agg['p2'].isin(pidx) &
    test_agg['p3'].isin(pidx)
].reset_index(drop=True)
test_known['in_train'] = test_known['lineId'].isin(set(train_agg['lineId']))

I_tst = np.zeros((len(test_known), n_pl), dtype=int)
y_tst_cf = np.zeros(len(test_known), dtype=float)
for i, (_, row) in enumerate(test_known.iterrows()):
    for p in [row['p1'], row['p2'], row['p3']]:
        I_tst[i, pidx[p]] = 1
    y_tst_cf[i] = row['cf_pct']
I_tst = I_tst[:, rcm]

y_tst_bin = (y_tst_cf > 0.5).astype(float)

good = [k for k in range(I_trn.shape[1]) if I_trn[:, k].max() == 1]
bad = [k for k in range(I_trn.shape[1]) if I_trn[:, k].max() == 0]
if len(bad) > 0:
    mask = np.ones(len(I_tst), dtype=bool)
    for k in bad:
        mask &= (I_tst[:, k] == 0)
    I_tst = I_tst[mask]
    y_tst_cf = y_tst_cf[mask]
    y_tst_bin = y_tst_bin[mask]
    test_known = test_known[mask].reset_index(drop=True)
I_trn = I_trn[:, good]
I_tst = I_tst[:, good]

known_mask = test_known['in_train'].values
new_mask = ~known_mask
majority_acc = max(y_tst_bin.mean(), 1 - y_tst_bin.mean())

print(f"{'='*65}")
print(f"ДАННЫЕ: I_trn={I_trn.shape}, I_tst={I_tst.shape}")
print(f"Majority: {majority_acc:.3f}")
print(f"{'='*65}")

# ============================================================
# Функция оценки модели
# ============================================================

def evaluate(name, y_pred, y_true_cf, y_true_bin, known_mask, new_mask):
    y_pred = np.clip(y_pred, 0, 1)
    acc = np.mean((y_pred > 0.5) == y_true_bin)
    mse = np.mean((y_pred - y_true_cf) ** 2)
    mae = np.mean(np.abs(y_pred - y_true_cf))
    try:
        auc = roc_auc_score(y_true_bin, y_pred)
    except:
        auc = 0.0
    acc_k = np.mean((y_pred[known_mask] > 0.5) == y_true_bin[known_mask]) if known_mask.sum() > 0 else 0
    acc_n = np.mean((y_pred[new_mask] > 0.5) == y_true_bin[new_mask]) if new_mask.sum() > 0 else 0
    return {'name': name, 'acc': acc, 'auc': auc, 'mse': mse, 
            'mae': mae, 'acc_k': acc_k, 'acc_n': acc_n, 'pred': y_pred}

results = []

# ============================================================
# Бейзлайн: Ridge
# ============================================================

ridge = Ridge(alpha=10.0)
ridge.fit(I_trn, y_trn)
pred_ridge = ridge.predict(I_tst)
res = evaluate('Ridge (α=10)', pred_ridge, y_tst_cf, y_tst_bin, known_mask, new_mask)
results.append(res)
print(f"\nRidge: Acc={res['acc']:.4f}, AUC={res['auc']:.4f}")

# ============================================================
# ANOVA order=1 + ALS (предыдущий лучший)
# ============================================================

print(f"\n{'='*65}")
print(f"ANOVA order=1, r=2 + ALS (λ=0.1):")
print(f"{'='*65}")

Y0_o1 = teneva.anova(I_trn, y_trn, r=2, order=1)
pred_anova1 = np.clip(teneva.get_many(Y0_o1, I_tst), 0, 1)
res_a1 = evaluate('ANOVA ord=1', pred_anova1, y_tst_cf, y_tst_bin, known_mask, new_mask)
results.append(res_a1)
print(f"  ANOVA only: Acc={res_a1['acc']:.4f}, AUC={res_a1['auc']:.4f}")

Y_o1 = teneva.als(
    I_trn, y_trn, Y0_o1,
    nswp=30, lamb=0.1,
    I_vld=I_tst, y_vld=y_tst_cf,
    e_vld=1e-3, log=True
)
pred_als1 = np.clip(teneva.get_many(Y_o1, I_tst), 0, 1)
res_als1 = evaluate('ANOVA1+ALS r=2', pred_als1, y_tst_cf, y_tst_bin, known_mask, new_mask)
results.append(res_als1)
print(f"  +ALS: Acc={res_als1['acc']:.4f}, AUC={res_als1['auc']:.4f}")

# ============================================================
# ANOVA order=2 (ПАРНЫЕ ВЗАИМОДЕЙСТВИЯ!)
# ============================================================

print(f"\n{'='*65}")
print(f"ANOVA order=2 — ПАРНЫЕ ВЗАИМОДЕЙСТВИЯ")
print(f"{'='*65}")

for r in [2, 3, 5]:
    print(f"\n--- r={r} ---")
    
    t = tpc()
    Y0_o2 = teneva.anova(I_trn, y_trn, r=r, order=2)
    t_anova = tpc() - t
    
    pred_anova2 = np.clip(teneva.get_many(Y0_o2, I_tst), 0, 1)
    res_a2 = evaluate(f'ANOVA ord=2 r={r}', pred_anova2, 
                       y_tst_cf, y_tst_bin, known_mask, new_mask)
    results.append(res_a2)
    print(f"  ANOVA only: Acc={res_a2['acc']:.4f}, AUC={res_a2['auc']:.4f}, "
          f"MSE={res_a2['mse']:.4f} ({t_anova:.1f}s)")
    
    # ALS с разными λ
    for lamb in [0.1, 1.0, 10.0]:
        t = tpc()
        Y_o2 = teneva.als(
            I_trn, y_trn, Y0_o2,
            nswp=30, lamb=lamb,
            I_vld=I_tst, y_vld=y_tst_cf,
            e_vld=1e-3, log=True
        )
        t_als = tpc() - t
        
        pred_als2 = np.clip(teneva.get_many(Y_o2, I_tst), 0, 1)
        pred_als2_trn = np.clip(teneva.get_many(Y_o2, I_trn), 0, 1)
        
        trn_acc = np.mean((pred_als2_trn > 0.5) == (y_trn > 0.5))
        
        name = f'ANOVA2+ALS r={r} λ={lamb}'
        res_als2 = evaluate(name, pred_als2, y_tst_cf, y_tst_bin, 
                            known_mask, new_mask)
        results.append(res_als2)
        
        print(f"  +ALS λ={lamb}: TrnAcc={trn_acc:.4f}, "
              f"TstAcc={res_als2['acc']:.4f}, AUC={res_als2['auc']:.4f}, "
              f"MSE={res_als2['mse']:.4f} ({t_als:.1f}s)")

# ============================================================
# Итоговая таблица
# ============================================================

print(f"\n{'='*80}")
print(f"ИТОГОВАЯ ТАБЛИЦА:")
print(f"{'='*80}")
print(f"{'Модель':<28s} {'Acc':>7} {'AUC':>7} {'MSE':>8} "
      f"{'MAE':>7} {'Знак':>7} {'Нов':>7}")
print(f"{'-'*80}")
print(f"{'Majority':<28s} {majority_acc:>7.4f}")

# Сортируем по AUC
results_sorted = sorted(results, key=lambda x: x['auc'], reverse=True)

for res in results_sorted:
    print(f"{res['name']:<28s} {res['acc']:>7.4f} {res['auc']:>7.4f} "
          f"{res['mse']:>8.4f} {res['mae']:>7.4f} "
          f"{res['acc_k']:>7.4f} {res['acc_n']:>7.4f}")

best = results_sorted[0]
print(f"\n  ★ Лучшая по AUC: {best['name']}, "
      f"Acc={best['acc']:.4f}, AUC={best['auc']:.4f}")

# ============================================================
# Визуализация
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1: Accuracy — order=1 vs order=2 vs Ridge
key_models = [r for r in results if any(k in r['name'] for k in 
              ['Ridge', 'ANOVA1+ALS', 'ANOVA2+ALS r=2', 'ANOVA2+ALS r=3'])]
names = [r['name'] for r in key_models]
accs = [r['acc'] for r in key_models]
colors_bar = ['gray' if 'Ridge' in n else 'steelblue' if 'ord=1' in n 
              else 'darkblue' for n in names]
axes[0,0].barh(range(len(names)), accs, color=colors_bar)
axes[0,0].set_yticks(range(len(names)))
axes[0,0].set_yticklabels(names, fontsize=8)
axes[0,0].axvline(majority_acc, color='red', linestyle='--', label='Majority')
axes[0,0].set_xlabel('Test Accuracy')
axes[0,0].set_title('Accuracy: order=1 vs order=2 vs Ridge')
axes[0,0].legend()

# 2: AUC
aucs_bar = [r['auc'] for r in key_models]
axes[0,1].barh(range(len(names)), aucs_bar, color=colors_bar)
axes[0,1].set_yticks(range(len(names)))
axes[0,1].set_yticklabels(names, fontsize=8)
axes[0,1].axvline(0.5, color='red', linestyle='--')
axes[0,1].set_xlabel('Test AUC')
axes[0,1].set_title('AUC: order=1 vs order=2 vs Ridge')

# 3: Predicted vs Actual — лучшая order=2
best_o2 = [r for r in results if 'ANOVA2' in r['name']]
best_o2 = max(best_o2, key=lambda x: x['auc']) if best_o2 else results[0]

axes[1,0].scatter(y_tst_cf, pred_ridge, alpha=0.4, s=20, 
                   label='Ridge', color='gray')
axes[1,0].scatter(y_tst_cf, best_o2['pred'], alpha=0.4, s=20, 
                   label=best_o2['name'], color='darkblue')
axes[1,0].plot([0.3, 0.7], [0.3, 0.7], 'r--')
axes[1,0].axhline(0.5, color='gray', linestyle=':')
axes[1,0].axvline(0.5, color='gray', linestyle=':')
axes[1,0].set_xlabel('Actual CF%')
axes[1,0].set_ylabel('Predicted CF%')
axes[1,0].set_title('Predicted vs Actual')
axes[1,0].legend()

# 4: Знакомые vs Новые
o1_res = res_als1
o2_res = best_o2
ridge_res = results[0]

x = np.arange(3)
w = 0.35
axes[1,1].bar(x - w/2, 
              [ridge_res['acc_k'], o1_res['acc_k'], o2_res['acc_k']], 
              w, label='Знакомые', color='steelblue')
axes[1,1].bar(x + w/2, 
              [ridge_res['acc_n'], o1_res['acc_n'], o2_res['acc_n']], 
              w, label='Новые', color='coral')
axes[1,1].set_xticks(x)
axes[1,1].set_xticklabels(['Ridge', 'TT ord=1', f'TT ord=2'])
axes[1,1].axhline(majority_acc, color='red', linestyle='--', label='Majority')
axes[1,1].set_ylabel('Accuracy')
axes[1,1].set_title('Знакомые vs Новые тройки')
axes[1,1].legend()

plt.suptitle('ANOVA order=2 (парные взаимодействия): сезон 2023',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('fig_anova_order2.png', dpi=150, bbox_inches='tight')
plt.show()
```
